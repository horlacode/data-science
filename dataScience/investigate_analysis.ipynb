```python
df1 = pd.read_csv('RealGDP08062020.csv')
df2 = pd.read_csv('inflations.csv')
df3 = pd.read_csv('exchanged.csv')
```

```python
#investigating how inflation and exchange rate affect GDP
#year as factor - start from 2004 - 2008
#for gdp df,take only TOTAL GDP column
#for exchange ,take only YUAN and USD

df1.head()
```

```python
gdp_2004_2008 = np.round(df1[(df1['tyear']>=2004) & (df1['tyear']<2009)]['TOTAL_GDP'],1
)

```

```python
df1[(df1['tyear']>=2004) & (df1['tyear']<2009)]
```

```python
inflation_2004_2008 = df2[(df2['Year']>=2004) & (df2['Year']<2009)]['All Items (Year on Change)']
```

```python
exchange_2004_2008 = df3[(df3['Rate Year']>=2004) & (df3['Rate Year']<2009)][['Rate Year','Currency','Central Rate']]
```

```python
df3['Currency'].unique()
```

```python
inflation_2004_2008
```

```python
#to summarise the row for every 12 months in a year
y2004 = np.round(np.mean(inflation_2004_2008.iloc[0:12]),1)
y2005 = np.round(np.mean(inflation_2004_2008.iloc[12:24]),1)
y2006 = np.round(np.mean(inflation_2004_2008.iloc[24:36]),1)
y2007 = np.round(np.mean(inflation_2004_2008.iloc[36:48]),1)
y2008 = np.round(np.mean(inflation_2004_2008.iloc[48:60]),1)

inflation = [y2004,y2005,y2006,y2007,y2008]
```

```python
inflation  
```

```python
exchange_2004_2008 = df3[(df3['Rate Year']>=2004) & (df3['Rate Year']<2009) & (df3['Currency']=='US DOLLAR')][['Rate Year','Currency','Central Rate']]

ex_2008 = np.round(np.mean(exchange_2004_2008[exchange_2004_2008['Rate Year']==2008]),1)
ex_2007 = np.round(np.mean(exchange_2004_2008[exchange_2004_2008['Rate Year']==2007]),1)
ex_2006 = np.round(np.mean(exchange_2004_2008[exchange_2004_2008['Rate Year']==2006]),1)
ex_2005 = np.round(np.mean(exchange_2004_2008[exchange_2004_2008['Rate Year']==2005]),1)
ex_2004 = np.round(np.mean(exchange_2004_2008[exchange_2004_2008['Rate Year']==2004]),1)

exchange = [ex_2004['Central Rate'],ex_2005['Central Rate'],ex_2006['Central Rate'],ex_2007['Central Rate'],ex_2008['Central Rate']]
```

```python
ex_2008
```

```python
#to create our own data frame
data = {'Year': [2004,2005,2006,2007,2008],
        'GDP': gdp_2004_2008,
        'Inflation':inflation,
        'USD Rate': exchange
        
        } #year,GDP,Inflation,USD RATE
df = pd.DataFrame(data)
```
