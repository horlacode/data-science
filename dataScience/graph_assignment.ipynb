```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('RealGDP08062020.csv')
```

```python
df
```

```python
df_81_01 = df[(df['tyear']>=1981) & (df['tyear']<=2001)]
df_81_01[['tyear','Agriculture','Industry','Forestry']]

year = df_81_01['tyear']

ag = df_81_01['Agriculture']/1000000
agric = ag.round(1)

ind = df_81_01['Industry']/1000000
industry = ind.round(1)

fo =  df_81_01['Forestry']/1000000
forestry = fo.round(1)
```

```python
plt.style.use('seaborn')





plt.plot(year,agric,color='g',marker='o',linewidth=3,label='Agriculture')


plt.plot(year,industry,color='c',marker='',linestyle='-.',linewidth=3,label='Industry')

plt.plot(year,forestry,color='k',linestyle='dashdot',linewidth=3,marker='.',label='Forestry')



plt.ylabel('Revenue(B)')
plt.xlabel('Ages')
plt.title('Total GDP (Naira) from Agriculture,Forestry and Industry Sector')

#plt.legend() is use to signal wat each of the lines represent there are two way of doing it

plt.legend()


plt.tight_layout()
plt.grid()

plt.savefig('assignment.pdf',dpi=600)

plt.show()
```
