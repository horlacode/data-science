```python
from matplotlib import pyplot as plt
```

```python
dex_x = [25,26,27,28,29,30]
dey_y = [30000,20000,10000,40000,50000,60000]

plt.plot(dex_x,dey_y,marker='*',color = '#444444',ls='--', label='Php')
plt.title('Income of programmers according to their age')

dey_p = [10000,20000,50000,30000,60000,40000]
plt.plot(dex_x,dey_p, marker='.',linewidth=3, label='Python')

dey_j = [60000,50000,30000,40000,10000,20000]
plt.plot(dex_x,dey_j,marker='o', label='Javascript')

plt.ylabel('Income($)')
plt.xlabel('Age')
plt.legend()
plt.show()
```
