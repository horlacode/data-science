```python
import matplotlib.pyplot as plt
import math as mt

import pandas as pd

%matplotlib inline
```

```python
gdp = pd.read_csv('RealGDP08062020.csv')
```

```python
gdp.head()
#real estate, financeand Insurance,education ,Public Adminstration
```

```python
#to get the data for real estate, financeand Insurance,education ,Public Adminstration ou
refp = gdp[['tyear','RealEstate','Education','FinanceAndInsurance','PublicAdministration']]
refp.head()
```

```python
#total for each industry
real_est = mt.floor(sum(refp['RealEstate']))/1000000
edu = mt.floor(sum(refp['Education']))/1000000
public = mt.floor(sum(refp['PublicAdministration']))/1000000
fin = mt.floor(sum(refp['FinanceAndInsurance']))/1000000


data = [real_est,edu,public,fin]
labels = 'Real Estate','Education','Finace','Pub Admin'
```

```python
labels
```

```python
#Bar chart 
plt.bar(labels,data)
plt.ylabel('Total Revenue')
plt.xlabel('Industry')
plt.title('Total Revenue 1981 - 2019')
plt.show()
```

```python
#bar chart using object oriented methods
fig,ax = plt.subplots()
plt.style.use('seaborn')

ax.bar(labels,data)
ax.set_xlabel('Industry')
ax.set_ylabel('Industry')
ax.set_title('Total Revenue 1981 - 2019')
plt.show()
ax.bar(labels,data)
ax.set_xlabel('Industry')
ax.set_ylabel('Industry')
ax.set_title('Total Revenue 1981 - 2019')
plt.show()
```

```python
#bar chart using object oriented methods
fig,ax = plt.subplots(figsize=(8,6))
ax.set_facecolor('orange')
ax.bar(labels,data, color='gray')
ax.set_xlabel('Industry')
ax.set_ylabel('Industry')
ax.set_title('Total Revenue 1981 - 2019')

plt.show()
```

```python
#pie chart
fig,ax2 = plt.subplots(figsize=(8,8))
ax2.pie(data,labels=labels,autopct='%1.1f%%')#to show the percent we do autopct='%1.1f%%'
plt.show()
```

```python
#to change the pie chart anti_clockwise we include startangle=90

fig,ax2 = plt.subplots(figsize=(8,8))
ax2.pie(data,labels=labels,autopct='%1.1f%%',startangle=90)#to show the percent we do autopct='%1.1f%%'
plt.show()
```

```python
#to make emphasis on a particular part u use explode
fig,ax2 = plt.subplots(figsize=(8,8))
ax2.pie(data,labels=labels,autopct='%1.1f%%',startangle=90,explode=(0.1,0.1,0.1,0.0),shadow=True)#to show the percent we do autopct='%1.1f%%'
plt.show()
```
