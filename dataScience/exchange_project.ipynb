```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('exchanged.csv')

#to replace 
data.replace('US DOLLAR','USD',inplace=True)
data.replace('POUNDS STERLING','GBP',inplace=True)
data.replace('SWISS FRANC','CHF',inplace=True)
data.replace('EURO','EUR',inplace=True)


#drop is use to remove from the table
data = data[(data['Currency']=='USD') & (data['Rate Year']==2019)]
data.drop('Currency',axis=1,inplace=True)
data.drop('Rate Year',axis=1,inplace=True)
data.drop('Buying Rate',axis=1,inplace=True)
data.drop('Selling Rate',axis=1,inplace=True)

data
```
