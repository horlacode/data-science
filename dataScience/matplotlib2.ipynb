```python
import matplotlib.pyplot as plt

import numpy as np

%matplotlib inline
```

```python
data1 = np.linspace(1,5,10)
data2 = data1 **2
data3 = data1 **3
```

```python
#Object Oriented method with matplotlib
#plotting a single line chart
fig = plt.figure(figsize=(10,5))
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(data1,data2,'r')


#set labels
axes.set_xlabel('data1 figures')
axes.set_ylabel('data2 figure')
axes.set_title('00 line plot')
```

```python
#plotting multiple charts on the same axes
fig = plt.figure(figsize=(10,5))
axes = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes.plot(data1,data2,'b', marker='s',markersize=8,markerfacecolor='r',markeredgewidth=3,markeredgecolor='r') 
#marker style: + ,*,s,p

#small chart
axes = fig.add_axes([0.2,0.4,0.3,0.4])
axes.plot(data1,data3,'b',ls='--',linewidth=3)

#to save as pictre
#plt.savefig('chart2.jpg' dpi=600) 
#dpi makes the image resolution better
```
