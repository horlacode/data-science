```python
n = pd.read_csv('exchange.csv')
n2 = n[(n['Rate Year']>2005)&(n['Rate Year']<2010)]
usd = n2[n2['Currency']=='US DOLLAR']

USD = usd[['Rate Year','Rate Month','Buying Rate']]
pv = USD.pivot_table(index='Rate Year',columns='Rate Month',values='Buying Rate')
```
